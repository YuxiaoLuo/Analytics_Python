```python
import pandas as pd

# create a dataframe
df = pd.DataFrame({'key':['A','B','C','A','B','C'],
                  'data':range(6)}, columns = ['key','data'])
df
```

```python
# simple GroupBy
df.groupby('key')
           
# operations on groupby 
df.groupby('key').sum()

df.groupby('key').mean()

print(type(df.groupby('key')))
```

```python
# aggregate, filter, transform, apply 
import numpy as np
import pandas as pd

# initialization of random seed in rng to use in randint 
rng = np.random.RandomState(0)
df = pd.DataFrame({'key':['A','B','C','A','B','C'],
                  'data1': range(6),
                  'data2':rng.randint(0,10,6)},
                   columns = ['key', 'data1', 'data2'])
df

# aggregation 

# aggregate can take a string, a function, or a list thereof 
# and compute all the aggregates at once 
df.groupby('key').aggregate(['min', np.median, max])

# you have options in doing string, function, or method: ['min','median','max']
df.groupby('key').aggregate({'data1':'min',
                             'data2':'max'})
```

```python
# Filtering
def filter_func(x):
    return x['data2'].std() > 4

print(df)
print(df.groupby('key').std())
print(df.groupby('key').filter(filter_func))
```

```python
# transformation 
def trans_func(x):
    return x - x.mean()

df.groupby('key').transform(trans_func)
```

```python
# Apply method 
def norm_by_data2(x):
    # x is a DataFrame of group values
    x['data1'] /= x['data2'].sum()
    return x
print(df)
print(df.groupby('key').apply(norm_by_data2))
```

```python
# Planets exercise 
import seaborn as sns
planets = sns.load_dataset('planets')

planets.head()

# dimesion of the dataset
planets.shape
```

```python
planets.groupby('method').sum()
planets.groupby('method')['number'].sum()
```

```python
planets.groupby('method')['orbital_period'].median()
```

```python
# iteration over groups
for (method, group) in planets.groupby('method'):
    print("{0:30s} shape  = {1}".format(method, group.shape))
```

```python
# Dispatch methods allows to call pandas methods on the groups 
planets.groupby('method')[['distance', 'number']].describe()
```

```python
# Challenge: count discovered planets by method and decade
# First part: create an additional column with the decades 

# divide each year by 10 and add a 0 at the end 
decade = 10*(planets['year']//10)
# cast decade as string
decade = decade.astype(str) + 's'
decade.name = 'decade'

decade
planets.shape

planets.groupby(['method',decade])['number'].sum().unstack().fillna('NA')

```
