```python
# !pip install spotipy
import spotipy
import spotipy.util as util
import json
import os
import sys
import spotipy.oauth2 as oauth2
import requests
import re
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="*****",
                                                           client_secret="*****"))
```

```python
def get_top_songs_2020(artist):
    # get credentials
    results = sp.search(q=artist, limit=20)
    top_songs = []
    #for idx, track in enumerate(results['tracks']['items']):
    for idx, track in enumerate(results['tracks']['items']):
        top_songs.append(track['name'])
    print(artist,top_songs)

    # get lyrics
    for i in range(len(top_songs)):
        #song = top_songs[i]
        #print(song)
        #response = requests.get('https://api.lyrics.ovh/v1/{}/{}'.format(artist,song))
        response = requests.get('https://api.lyrics.ovh/v1/{}/{}'.format(artist,i))
        print(response)
        if response.status_code == 200:
            lyrics = response.json()
            lyrics = lyrics['lyrics']
            lyrics = re.sub(r'\[.*\]', '', lyrics)
            lyrics = re.sub(r'\n', ' ', lyrics)
            lyrics = re.sub(r'\r', '', lyrics)
            lyrics = re.sub(r'\t', '', lyrics)
            lyrics = re.sub(r'  ', '', lyrics)
            print(lyrics)
        else:
            return
```
