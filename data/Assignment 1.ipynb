```python
#question5
import requests 
url ="https://en.wikipedia.org/wiki/Baruch_College"
r= requests.get(url)
html_contents=r.text

from bs4 import BeautifulSoup 

html_soup = BeautifulSoup (html_contents)

allExtern= html_soup.find_all('a')
allpara=html_soup.find_all('p')
allTable = html_soup.find_all ('Table')

cntExtern , cntPara , cnTable=0,0,0

for x in allExtern:
    cntExtern +=1
for x in allpara:
    cntPara +=1
for x in allTable:
    cntTable +=1
    
print (f"External links: {cntExtern}\nParagraphs: {cntPara}\nTables:{cnTable}")
```
